### Line Plot

```python
import plotly.graph_objects as go

df_stocks = px.data.stocks()
px.line(df_stocks, x='date', y='GOOG', labels={'x': 'date', 'y': 'price'})
```

```python
px.line(df_stocks, x='date', y=['GOOG', 'AAPL'], labels={'x': 'date', 'y': 'price'}, title="Stock")
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=df_stocks['date'], y=df_stocks['GOOG'],
                         mode='lines', name='Google'))

fig.add_trace(go.Scatter(x=df_stocks['date'], y=df_stocks['AAPL'],
                         mode='lines', name='Apple'))

fig.add_trace(go.Scatter(x=df_stocks['date'], y=df_stocks['MSFT'],
                         mode='lines+markers', name='Microsoft'))

fig.add_trace(go.Scatter(x=df_stocks['date'], y=df_stocks['AMZN'],
                         mode='lines', name='Amazon', line=dict(color='firebrick')))

fig.update_layout(
    xaxis=dict(showline=True, showgrid=True, showticklabels=True, linecolor='rgb(12, 34, 56)'),
)
```
